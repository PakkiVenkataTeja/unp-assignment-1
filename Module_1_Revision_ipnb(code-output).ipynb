```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("Housing.xlsx",index_col=0)
```

```python
data.head()
```

```python
np.sum(data.isna())
```

```python
for i in data.columns:
    if (i !="PropertyAddress" and i != 'LandUse' and i != 'SaleDate' and i != 'SalePrice'):
        data[i].fillna(np.mean(data[i]),inplace=True)
```

```python
data
```

```python
np.sum(data.isna())
```

```python
data['LandValue'].round(decimals = False)
```

```python
data.rename(columns={"PropertyAddress":"Location"},inplace=True)
data.rename(columns={"SalePrice":"Cost"},inplace=True)
```

```python
data
```

```python
data['year'] = pd.DatetimeIndex(data['SaleDate']).year
data['month'] = pd.DatetimeIndex(data['SaleDate']).month
data['pin']=data['Location'].str[:4]
```

```python
##data['sorted_yr'] = sorted(data['year'])
data['lvr'] = data['LandValue'].round(decimals = False)
```

```python
data
```

**1. How has the price of properties changed over the years**

From 2013 - 2015 price of the properties increased rapidly. In between 2015 price of the properties started to decrease and there is no sales of property in year 2019 expert in the month of May and December.

```python
fig, axs = plt.subplots(nrows=2) 
fig.set_size_inches(15, 20);
sns.lineplot(x="year",y="Cost",data=data,ax=axs[0])
sns.barplot(x="year",y="Cost",data=data,ax=axs[1])
```

```python
data.groupby("year")["Cost"].sum()
```

**2. Does a specific month or span of months have the highest property prices?**

In the year 2015 1st month ( 1541095929 ) have the highest property value.

```python
fig, axs = plt.subplots(nrows=2) 
fig.set_size_inches(15, 20);
sns.barplot(x="year",y="Cost",data=data, hue = "month",ax=axs[0])
sns.barplot(x="month",y="Cost",data=data,hue="year",ax=axs[1])
```

```python
data.groupby(["year","month"])["Cost"].sum()
```

**7. Does the size of the property (no. of acres) have any influence in its price?**

The properties which are in between 33 - 41 acres have highest price where as properties which are in between 0.01 - 13 acres have average price with more number of plots.
